```python
from math import sin, exp

f = lambda x: sin(x/5.0) * exp(x/10.0) + 5 * exp(-x/2.0)
def fu(x):
    return sin(x/5.0) * exp(x/10.0) + 5 * exp(-x/2.0)
print f(1), ' ', f(15)
```

```python
import numpy as np
from scipy.linalg import solve
from matplotlib import pylab as plt
```

```python
b = np.array([fu(1), fu(15)])
a = np.array([[1, 1], [1, 15]])
w = solve(a, b)
print(["%.2f" % x for x in w])
x = np.arange(1, 16)
yp = w[0] + w[1] * x
yf = [fu(y) for y in x]

plt.plot(x, yp)
plt.plot(x, yf)
plt.show()

```

```python
b = np.array([fu(1), fu(8), fu(15)])
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
w = solve(a, b)
print(["%.2f" % e for e in w])

yp = w[0] + w[1] * x + w[2] * (x ** 2)

plt.plot(x, yp)
plt.plot(x, yf)
plt.show()
```

```python

b = np.array([fu(1), fu(4), fu(10), fu(15)])
a = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 225*15]])

w = solve(a, b)
print(["%.2f" % e for e in w])


yp = w[0] + w[1] * x + w[2] * (x ** 2) + w[3] * (x ** 3)


plt.plot(x, yp)
plt.plot(x, yf)
plt.show()
```
